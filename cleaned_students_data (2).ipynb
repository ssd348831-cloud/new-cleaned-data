```python
import json
```

```python
json_file=r"C:\Users\titio\Downloads\messy_students_data.json"
```

```python
# Open and load JSON file
with open(json_file, "r") as file:
    data = json.load(file)

# Loop through every student
for student in data:

    # Clean the data
    name = student["full_name"].title().strip()
    email = student["emailAddress"].lower().strip()
    score = int(student["score"])
    department = student["department"].title().strip()

    # Handle missing status
    status =  "Pass" if score >= 50 else "Fail"

    # Print cleaned data
    print("Name:", name)
    print("Email:", email)
    print("Score:", score)
    print("Department:", department)
    print("Status:", status)
    print("-" * 30)
    
```
