```python
df = pd.DataFrame({
    'full_name': ['  taiwo ade', 'DeMi LADE', ' aRa OWO', 'Joseph PEP  '],
    'emailAddress': ['TAIWO@yahoo.COM', 'DEMMY@outlook.com ', 'ARAOWO@gmail.com', ' PEPG@hotmail.com '],
    'score': ['82', '91 ', '45', '77'],
    'department': [' engineering', 'SCIENCE ', ' arts ', 'Sports']
})

# Show original data
print("Original Data:")
print(df)
print("\n" + "="*60 + "\n")

# Clean the data (updated column names to match your data)
df['full_name'] = df['full_name'].str.title().str.strip()
df['emailAddress'] = df['emailAddress'].str.lower().str.strip()
df['score'] = df['score'].str.strip().apply(lambda x: int(x))  # Added str.strip() to handle spaces
df['department'] = df['department'].str.title().str.strip()

# Add Pass/Fail status
df['Status'] = df['score'].apply(lambda x: "Pass" if x >= 50 else "Fail")

# Show cleaned data
print("Cleaned Data:")
print(df)

# Save file
df.to_csv('cleaned_students_data.csv', index=False)
```
