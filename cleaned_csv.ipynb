```python
df = pd.DataFrame({
    'Name': ['john doe', 'jane smith', 'bob johnson'],
    'EMAIL': ['JOHN@EMAIL.COM', 'JANE@EMAIL.COM', 'BOB@EMAIL.COM'],
    'Score': ['75', '45', '88'],
    'Department': ['engineering', 'marketing', 'sales']
})

# Show original data
print("Original Data:")
print(df)
print("\n" + "="*50 + "\n")

# Your original code will now work
df['Name'] = df['Name'].str.title().str.strip()
df['EMAIL'] = df['EMAIL'].str.lower().str.strip()
df['Score'] = df['Score'].apply(lambda x: int(x)) # <-- Changed here
df['Department'] = df['Department'].str.title().str.strip()

# Add Pass/Fail status
df['Status'] = df['Score'].apply(lambda x: "Pass" if x >= 50 else "Fail")

# Show cleaned data
print("Cleaned Data:")
print(df)

# Save file
df.to_csv('cleaned_students_data.csv', index=False)
print("\nFile saved as 'cleaned_students_data.csv'")
```
